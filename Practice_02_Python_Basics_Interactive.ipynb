## 1) Variables and Types

A variable is a name that points to a value.
Python decides the type **at runtime**.

```python
gene = "TP53"
reads = 42
ph = 7.4
is_coding = True

print(gene, type(gene))
print(reads, type(reads))
print(ph, type(ph))
print(is_coding, type(is_coding))
```

## 2) Dynamic Typing

The **same variable name** can later point to a value of a different type.

```python
x = 5
print(x, type(x))

x = "five"
print(x, type(x))
```

## 3) Arithmetic Operators (Math)

Here are the main arithmetic operators:

- `+` addition
- `-` subtraction
- `*` multiplication
- `/` division (returns a float)
- `//` floor division
- `%` remainder
- `**` power

### Predict before running

What will each line print?
- `10 / 3`
- `10 // 3`
- `10 % 3`
- `2 ** 3`

```python
a = 10
b = 3

print("a / b  =", a / b)
print("a // b =", a // b)
print("a % b  =", a % b)
print("2 ** 3 =", 2 ** 3)
```

### Order of operations

Like in math, Python does **not** always calculate from left to right.
Multiplication happens before addition.

```python
print(2 + 3 * 4)
print((2 + 3) * 4)
```

## 4) Comparison Operators

Comparison operators return `True` or `False`.

- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to

### Predict before running

What do you expect from each comparison?

```python
x = 8
y = 12

print(x == y)
print(x != y)
print(x < y)
print(x >= y)
```

## 5) Logical Operators

Logical operators combine conditions:

- `and` -> both conditions must be `True`
- `or` -> at least one condition must be `True`
- `not` -> reverses `True` / `False`

```python
enough_reads = True
good_quality = False

print(enough_reads and good_quality)
print(enough_reads or good_quality)
print(not good_quality)
```

## 6) Strings

Strings are sequences of characters.
In biology, strings are often useful for:
- DNA sequences
- RNA sequences
- gene names
- sample IDs

```python
seq = "ATGC"
gene_name = "BRCA1"

print(seq, type(seq))
print(gene_name, type(gene_name))
```

### String operators

## 7) Indexing and Slicing

Python uses **0-based indexing**.

For the string `"HELLO"`:
- index `0` is `'H'`
- index `1` is `'E'`
- index `-1` is the last character

```python
word = "HELLO"

print(word[0])
print(word[1])
print(word[-1])
```

### Predict before running

What do these slices return?

```python
organism = "Streptococcus"

print(organism[0:4])
print(organism[:6])
print(organism[6:])
print(organism[::2])
print(organism[::-1])
```
