## 1. Functions and return values

A function can receive input, perform a calculation, and **return** a value.


```python
def fahr2cel(fahr_temp):
    '''Convert temperature from Fahrenheit to Celsius'''
    cel_temp = (fahr_temp - 32) / 1.8
    return cel_temp

print(fahr2cel(98.6))
print(fahr2cel(32))
```

## 3. Conditions as scientific rules

Programs can encode decision rules.


```python
def state(Tm, Tv, T):
    '''Return the state of a material based on melting/boiling points'''
    if T < Tm:
        return "solid"
    elif T < Tv:
        return "liquid"
    else:
        return "gas"

for T in [-10, 0, 20, 99, 100, 120]:
    print(T, "->", state(0, 100, T))
```

## 4. GC content – first implementation

This version scans the sequence one nucleotide at a time.


```python
def GC_content(dna_string):
    '''calculate %GC in DNA string (expects uppercase ACTG)'''
    GC = 0
    for nuc in dna_string:
        if nuc == "C" or nuc == "G":
            GC += 1
    return 100 * GC / len(dna_string)

print(GC_content("ATGCGC"))
print(GC_content("AAAAAA"))
```

## 5. GC content – alternative implementation

This version uses the string method `.count()`.


```python
def GC_content2(dna_string):
    '''calculate %GC in DNA string (expects uppercase ACTG)'''
    C = dna_string.count("C")
    G = dna_string.count("G")
    return 100 * (C + G) / len(dna_string)

print(GC_content2("ATGCGC"))
print(GC_content2("AAAAAA"))
```

## 7. Protein composition

This function measures the percentage of basic amino acids in a protein.


```python
def basic_aa_perc(prot):
    '''return percentage of basic amino acids in protein'''
    basic = 0
    basic_aa_list = ["K", "H", "R"]
    for aa in prot:
        if aa in basic_aa_list:
            basic += 1
    return 100 * basic / len(prot)

for prot in ["MKRH", "AAAA", "KRAHDDD", "RRRHHHKKKAAA"]:
    print(prot, "->", basic_aa_perc(prot))
```

### Think about the abstraction

This function is really asking:
**what fraction of the protein belongs to a chosen amino-acid class?**

## 8. Searching sequence motifs

This function finds **all** positions of a pattern in a sequence.


```python
def cut_positions(seq, pattern):
    '''find all positions of pattern within seq'''
    cut_sites = []
    site = seq.find(pattern)
    while site != -1:
        cut_sites.append(site)
        site = seq.find(pattern, site + 1)
    return cut_sites

print(cut_positions("AAAAAA", "AA"))
print(cut_positions("ATATAT", "ATA"))
print(cut_positions("GATATATGCATATACTT", "ATAT"))
```

## 9. Finding stop codons

This function scans a DNA sequence in frame 0 and returns the index of the first in-frame stop codon.


```python
def find_stop(dna):
    stops = ["TAG", "TAA", "TGA"]
    n = len(dna)
    for i in range(0, n, 3):
        if dna[i:i + 3] in stops:
            return i

print(find_stop("ATGAAATAGCCC"))
print(find_stop("AATGAAATAGCCC"))
print(find_stop("ATGAAATA"))
```
